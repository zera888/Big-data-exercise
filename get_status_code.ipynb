```python
# get.py
import requests
url='http://www.ehappy.com.tw/demo.htm'
html = requests.get(url)
# 檢查http回應碼是否為200(requests.code.ok)
if html.status_code == requests.codes.ok:
    print(html.text)
```

```python
# get_params.py
import requests
# 將查詢參數定義為字典資料加入get請求中
payload = {'key1': 'value1', 'key2': 'value2'}
html = requests.get("http://httpbin.org/get", params = payload)
print(html.text)
```

```python
# post.py
import requests
#將查詢參數加入post請求中
import requests
# 將查詢參數定義為字典資料加入get請求中
payload = {'key1': 'value1', 'key2': 'value2'}
html = requests.post("http://httpbin.org/post", data = payload)
print(html.text)
```

```python
# get_headers.py
import requests
url = 'http://irs.thsrc.com.tw/IMINT/'
#自訂標頭
headers={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:101.0) Gecko/20100101 Firefox/101.0'
}

#定表頭加入GET請求中
html = requests.get(url, headers=headers)
print(html)
```

```python
# get_cookie.py
url = 'http://www.ptt.cc/bbs/Gossiping/index.html'
#設定cookies的值
cookies = { 'over18' : '1'}
r = requests.get(url, cookies = cookies)
print(r.text)
```
